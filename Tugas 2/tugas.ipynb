```python
import pandas as pd
# Path file
file_path = "Presentase Penyelesaian Tindak Pidana di Indonesia tahun 2021-2022.csv"

# Membaca file CSV
df = pd.read_csv(file_path)

# Menampilkan beberapa baris pertama untuk melihat strukturnya
df.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Mengatur ukuran figure
plt.figure(figsize=(12, 6))

# Membuat bar plot
df_melted = df.melt(id_vars=["Kepolisian Daerah"], 
                     value_vars=["Penyelesaian tindak pidana 2021(%)", "Penyelesaian tindak pidana 2022(%)"],
                     var_name="Tahun", value_name="Persentase Penyelesaian")

# Mengganti label tahun untuk lebih mudah dipahami
df_melted["Tahun"] = df_melted["Tahun"].replace({
    "Penyelesaian tindak pidana 2021(%)": "2021",
    "Penyelesaian tindak pidana 2022(%)": "2022"
})

# Plot menggunakan seaborn
sns.barplot(data=df_melted, x="Kepolisian Daerah", y="Persentase Penyelesaian", hue="Tahun")

# Menyesuaikan tampilan
plt.xticks(rotation=90)  # Memiringkan label sumbu X agar terbaca
plt.xlabel("Kepolisian Daerah")
plt.ylabel("Persentase Penyelesaian Tindak Pidana (%)")
plt.title("Perbandingan Persentase Penyelesaian Tindak Pidana 2021 vs 2022")
plt.legend(title="Tahun")
plt.grid(axis="y", linestyle="--", alpha=0.7)

# Menampilkan plot
plt.show()
```
